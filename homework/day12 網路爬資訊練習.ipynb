```python
#Ettoday 網路爬蟲實作練習
#能夠利用 Request + BeatifulSour 撰寫爬蟲，並存放到合適的資料結構

import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.ettoday.net/news/news-list.htm'
#  1.取出今天所有的發文
ra=requests.get(url)
re=ra.text
soup=BeautifulSoup(re)
```

```python
title_tags=soup.find( class_ ="part_list_2").find_all('h3') #  注意 class 要加  _ (底線)     這邊用 'h3' 而不用'a'  a資料比較少                 
for x in title_tags:   #一個h3 完之後 會跳行 接另外一個 比較整齊了一些
    print(x.find(class_="date").text  ,  x.find('a').text )  #  .find_all('a')[-1].text
# 這邊 .text 會把標籤去掉 只剩下時間   find_all 就在 <a href 前面多了 [ , 所以阿 會變成 陣列哦 如果要 .text 要加[0] OR [-1]
#用   .find_all('a')[-1].text  原因
#有可能會有廣告出現，廣告也會用 <a> 的方式存在
#有時候會抓到的是廣告的 <a>。所以，這邊改成抓「最後一個 <a>」，不管有沒有廣告都會抓到正確的。

# 關於 find_all() 與  find() 的差異性
#在於一個會回傳 HTML 物件，一個會回傳包含 HTML 物件所組成的 List，
#因此 find('a') 會等同於 find_all('a')[0] 的結果。
#大部分情況，如果條件下得夠好，可以採用 find 就好，
#但有些時候條件不好下，只能用某個條件中的第 N 個的規則找到，就非得改用 find_all 再加上位置。
#另外 find_all 也會用在資料有規則的情況，例如我們想要抓所有的新聞（每一則新聞的 HTML 是類似的），我們也會採用 find_all。
```
