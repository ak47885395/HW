# 根據範例，取出今天所有的新聞

```python




browser = webdriver.Chrome("chromedriver.exe")
# Chrome  C要大寫
browser.get("https://www.ettoday.net/news/news-list.htm")


##
#  先設定 今天的日期
##
today_timeArray =  time.localtime(time.time())    #time.time() 可以傳回從 1970/1/1 00:00:00 算起至今的秒數  可以不加
#print(timearray)  struct_time 格式

#struct_time 格式的好處就是可以直接取出日期或時間裡面的任意數值
time_otherstyle =  time.strftime('%Y %m %d %H:%M:%S'  , today_timeArray)

print(time_otherstyle)  #2020 08 13 21:20:06


#  滑到 要的地方
while(1):
        html_source = browser.page_source
        
        
        soup=BeautifulSoup(html_source)
        
        final_title= soup.find(class_="part_list_2").find_all('h3')[-1]
        
        update_struct_time =  time.strptime ( final_title.find(class_="date").text   ,  "%Y/%m/%d %H:%M"  )
        
    
        if today_timeArray.tm_mday != update_struct_time.tm_mday:
            
            break
# print(soup.find(class_="part_list_2").find_all('h3').text)   因為是一個 list 不能直接 .text 直接print
    
       
        time.sleep(2)
         
        browser.execute_script("window.scrollTo(0 , document.body.scrollHeight)")

for i in soup.find(class_="part_list_2").find_all("h3"):
    
    element_struct_time = time.strptime( i.find(class_="date").text ,  "%Y/%m/%d %H:%M")
    ### 
   ##     extract the time of each news 
    ###
    if today_timeArray.tm_mday ==   element_struct_time.tm_mday:   #  是今天的才印
        print(i.text )
'''
1.要求 今天   使用 strptime 取出 時間格式  再取出今天的  數字

2. 迴圈內不斷  更新 取得 更下面的 資料  同時也要確保在滑下去之前  檢查一下 最後一個是不是 今天的新聞 如果不是要停下

3.取出的資料因為裡面含有  不是今天的新聞 ((因為滑下去不太可能剛好最後一個還是今天的

4. 根據 3.  在一段一段新聞 檢查是不是今天的



'''
        
```

# 取出現在時間兩小時內的新聞

```python
within_two_hour = 2*60*60
time_back_two_hour = time.localtime(time.time()-within_two_hour)

print(time.strftime('%Y %m %d %H:%M:%S'  , time_back_two_hour  ))


browser = webdriver.Chrome("chromedriver.exe")

browser.get("https://www.ettoday.net/news/news-list.htm")


while(1):
    
            html = browser.page_source
        
            soup = BeautifulSoup(html)
            
            print("sucess")
            
            last = soup.find(class_="part_list_2").find_all("h3")[-1]
            
            last_time =  time.strptime(  last.find(class_="date").text  ,  "%Y/%m/%d %H:%M")
            
                
            
            time.sleep(2)
            
            browser.execute_script("window.scrollTo( 0 ,  document.body.scrollHeight  )")
            
            if (last_time.tm_hour <= time_back_two_hour.tm_hour): #& (element_time.tm_min < time_back_two_hour.tm_min) :
                
                break     #####
                              #####   無法確認 滑下去 最後一個 小時是多少  只能 <= 就結束
                 ############################
            
for  i  in soup.find(class_="part_list_2").find_all("h3"):
    
    #  逐項比對時間
    
    element_time = time.strptime( i.find(class_="date").text ,  "%Y/%m/%d %H:%M" )
    
  #  if elemnt_time.tm_sec < time_back_two_hour.tm_sec:
    
    
 #   if element_time.tm_hour != time_back_two_hour.tm_hour :
          
    if  (element_time.tm_hour >= time_back_two_hour.tm_hour) :
        
        if (element_time.tm_hour == time_back_two_hour.tm_hour) & (element_time.tm_min < time_back_two_hour.tm_min):
        
            break
        
        else:
        
            print(i.text)   

                    
```

# 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.support.ui import Select
browser = webdriver.Chrome(executable_path="chromedriver.exe")

browser.get("https://www.ettoday.net/news/news-list.htm")


day_back = 3*24*60*60

day_back_time = time.localtime(time.time()-day_back)

print(time.strftime("%Y/%m/%d %H:%M"  , day_back_time  ))

### 快速跳躍
## 打開選單

mon_select = Select(browser.find_element_by_id("selM"))

# 選擇日期  月份

mon_select.select_by_value(str(day_back_time.tm_mon))


day_select = Select(browser.find_element_by_id("selD"))

day_select.select_by_value(str(day_back_time.tm_mday) )

### 選好 案查尋囉

browser.find_element_by_id("button").click()
###  其實可以不用   但是會重新刷新
```
