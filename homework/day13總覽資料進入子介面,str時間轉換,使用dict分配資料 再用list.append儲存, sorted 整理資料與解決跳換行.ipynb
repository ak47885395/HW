```python
#1.印出最新文章的「作者」「標題」「時間」
title=soup.find(class_="title").text.replace('\t','').replace('\n','')    #記得find 出來的有標籤 要加text
#new_article_url=soup.find( ('a')['href'].text , 'html5lib')  
#   不能直接 soup 來找  要變成 會改變的迴圈 一串一串 去find  所以往  a href 上一層看
for k in soup.find_all(class_="title"):
#    newest_article=k.find(  ('a')['href']  ).text    好吧不能存 要加在 網址後面
    r = BeautifulSoup(requests.get('https://www.ptt.cc'+k.find('a')['href']).text, "html5lib")
    time=r.find_all(class_='article-meta-value')[-1].text  #r.find_all(class_='article-meta-value')[-1].text
   # time = datetime.strptime(time , "%a %b %d %H:%M:%S %Y") # post['時間'] = datetime.strptime(post['時間'], "%a %b %d %H:%M:%S %Y") 
    author=r.find(class_='article-meta-value').text
    if 1:
        break
#print(r)
#print(title)
print(author , title , time)
print(type(time))
```

```python
#2.印出第一頁所有文章的「作者」「標題」「時間
posts=[]  #  這邊用list 因為最終 要 一串一串 的資料  
for d in soup.find_all(class_="title"):
    try:
        post={}  # 這邊用 dict 因為 一串資料中有 作者 標題  時間
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        post['作者']=r.find( class_="article-meta-value").text  #這邊原始碼是第0個
    
        post['標題']=d.text.replace('\t','').replace('\n','')  #這邊用d  d是總覽 已經有完整標題了
    
        post['時間']=  r.find_all(class_="article-meta-value")[-1].text# class_='article-meta-value' 有好幾個 用findall 往下找
        post['時間'] = datetime.strptime(post['時間'], "%a %b %d %H:%M:%S %Y") #這邊時間轉換一下
    
        posts.append(post)  #這裡一定要append 一下 不然 下個迴圈 post 會不見
    except:
            continue
print(posts) 
        
#    except:
#       continue   # 這會導致繞過錯誤 而沒有全部我們想要ˋ的結果
```

```python
posts = sorted(posts, key= lambda x: x['作者'])   # key 用在包著dict 的list 要排序時
#posts[3]
posts
# 另posts 這個list 中 最外層的每個元素叫做 x
#然後將每個 x 指定一個比較值，這個值來自於自己那個 x  裡面index= '時間' 的這一項作為比較標準。
#例如  作者 標題 時間,  key被指定為 時間。
```

```python
#  3. 試著爬爬看其他版的文章
#1.印出最新文章的「作者」「標題」「時間」


import requests
from bs4 import BeautifulSoup


re = requests.get("https://www.ptt.cc/bbs/Lifeismoney/index.html" ,  cookies = {'over18':'1'})

soup = BeautifulSoup(re.text)

subpage = soup.find_all(class_="r-ent")[0].a['href'] # the newsest content

resp = requests.get("https://www.ptt.cc"+subpage  ,  cookies = {'over18':'1'} )

soup = BeautifulSoup(resp.text)

main_container  = soup.find(id="main-content").find_all(class_="article-meta-value")

print('作者:' + main_container[0].text )

print('標題:' + main_container[2].text )

print('時間:' + main_container[-1].text)
y = datetime.strptime(main_container[-1].text)
print(y)
```

```python
#2.印出第一頁所有文章的「作者」「標題」「時間

re = requests.get("https://www.ptt.cc/bbs/Lifeismoney/index.html" ,  cookies = {'over18':'1'})

soup = BeautifulSoup(re.text)

thislist=[]
for i in soup.find_all(class_="r-ent"):
    
    if not i.a['href'] in thislist:
        thislist.append(i.a['href'])
for i in range(0,len(thislist)):
    subpage =  thislist[i]# the newsest content

    resp = requests.get("https://www.ptt.cc"+subpage  ,  cookies = {'over18':'1'} )

    soup = BeautifulSoup(resp.text)

    main_container  = soup.find(id="main-content").find_all(class_="article-meta-value")

    print('作者:' + main_container[0].text )

    print('標題:' + main_container[2].text )

    print('時間:' + main_container[-1].text)

```
